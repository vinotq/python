```python
class Passanger:
    def __init__(self, name: str, from_where: str, target: str):
        self.name = name
        self.from_where = from_where
        self.target = target

class Luggage:
    weight_limit = 47.0

    def check_weight(self, value):
        if not isinstance(value, float | int):
            raise TypeError('Weight must be float or int!')

        if value > self.weight_limit:
            raise ValueError('You can\'t exceed waight limit!')

        if value < 0:
            raise ValueError('Weight can\'t be negative!')

        return True

    def __init__(self, passanger: Passanger, weight: float | int):
        self.passanger = passanger
        if self.check_weight(weight):
            self.__weight = weight

    @property
    def weight(self):
        return self.__weight

    @weight.setter
    def weight(self, value):
        if self.check_weight(value):
            self.__weight = value

    def __str__(self):
        return f'Luggage of passanger {self.passanger.name} with weight {self.weight}'


class Cargo:
    def __init__(self):
        self._luggage_list = []

    def add(self, luggage: Luggage):
        if not isinstance(luggage, Luggage):
            raise TypeError('You can add only luggage!')

        self._luggage_list.append(luggage)

    def remove(self, luggage: Luggage):
        if not isinstance(luggage, Luggage):
            raise TypeError('You can remove only luggage!')

        self._luggage_list.remove(luggage)

    def get_all(self):
        return self._luggage_list

    def get_by_passanger(self, passanger: Passanger):
        return [luggage for luggage in self._luggage_list if luggage.passanger == passanger]

    @property
    def total_weight(self):
        return sum([luggage.weight for luggage in self._luggage_list])



passanger = Passanger('Oleg', 'Moscow', 'Afganistan')
print(passanger.name)
passanger2 = Passanger('Banan', 'Sirius', 'Albania')

luggage = Luggage(passanger, 10)
print(luggage)
luggage2 = Luggage(passanger, 47)
print(luggage2)
# luggage3 = Luggage(passanger, 52)
# print(luggage3)
luggage4 = Luggage(passanger, 32.45)
print(luggage4)

luggage21 = Luggage(passanger2, 15)
# luggage21.weight = -15
luggage21.weight = 26

cargo = Cargo()
print(cargo.total_weight)

cargo.add(luggage)
cargo.add(luggage2)

print([str(lug) for lug in cargo.get_all()])

cargo.add(luggage21)
print([str(lug) for lug in cargo.get_all()])

print(cargo.total_weight)

# cargo.remove(luggage)
# print([str(lug) for lug in cargo.get_all()])

print(cargo.total_weight)

print([str(lug) for lug in cargo.get_by_passanger(passanger)])
print([str(lug) for lug in cargo.get_by_passanger(passanger2)])

```
