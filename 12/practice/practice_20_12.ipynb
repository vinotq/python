Напишите описанную ниже систему классов и продемонстрируйте их работу:

Класс «Студент»:

Свойства класса:

Диапазон возможных оценок: tuple[int, int].

Свойства объекта:

- name (имя): str;
- average_mark (средний балл): float - read-only property;
- marks (оценки): list[int] - список оценок студента. Является property, сеттер проверяет тип данных оценки и вхождение в диапазон.

Методы:

- add_mark - добавляет новую оценку к оценкам студента;
- метод строкового представления, включающий в себя имя и средний балл.

Класс «Группа»:

Свойства:

- name (имя): str
- students: list[str] - read-only property.

Методы:

- Удаление и добавление студентов.
- У группы тоже есть read-only property average_mark, отображающее средний балл студентов группы.


```python
class Student:
    range_marks = (2, 5)

    def __init__(self, name: str):
        self.name = name
        self.__marks = []

    @property
    def marks(self):
        return self.__marks.copy()

    @marks.setter
    def marks(self, new_mark):
        for i in new_mark:
            if not isinstance(i, int):
                raise TypeError

            if not self.range_marks[0] <= i <= self.range_marks[-1]:
                raise ValueError

            self.__marks = new_mark

    @property
    def avg_mark(self):
        if len(self.__marks) == 0:
            print("There aren't any marks! DIVISION BY ZERO!")
            raise ZeroDivisionError
        else:
            return sum(self.__marks) / len(self.__marks)

    def add_mark(self, mark):
        if not isinstance(mark, int):
            raise TypeError

        if not self.range_marks[0] <= mark <= self.range_marks[-1]:
            raise ValueError

        self.__marks.append(mark)

    def __str__(self):
        return f"Student {self.name} has and average mark {self.avg_mark}"


class Group:
    def __init__(self, name: str, new_student):
        self.name = name
        self.__students = new_student

    @property
    def student(self):
        return [i.name for i in self.__students]

    def add_student(self, student):
        if not isinstance(student, Student):
            raise TypeError

        self.__students.append(student)

    def delete_student(self, student):
        if not isinstance(student, Student):
            raise TypeError
        self.__students.remove(student)

    @property
    def avg_mark_in_group(self):
        return sum([i.avg_mark for i in self.__students]) / len(self.__students)


stdnt = Student("Oleg")
stdnt.marks = [2, 3, 5]
print(stdnt.marks)

stdnt2 = Student("Bananov")
stdnt2.marks = [5, 5, 5]
print(stdnt2.marks)

stdnt.add_mark(3)
print(stdnt.marks)

stdnt2.add_mark(5)
print(stdnt2.marks)

print(stdnt)

print(stdnt2)

grp = Group("K0709-23/1", [stdnt, stdnt2])
print(grp.student)

stdnt3 = Student("Lyz")
stdnt3.marks = [4, 3, 4]

grp.add_student(stdnt3)
print(grp.student)

grp.delete_student(stdnt)
print(grp.student)

grp.add_student(stdnt)
print(grp.student)

print(grp.avg_mark_in_group)
```

```python
class Person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age


class Student(Person):
    def __init__(self, name: str, age: int, year=1):
        super().__init__(name, age)
        self.knowledge = 0
        self.year = year

    def get_knowledge(self):
        self.knowledge += 1


class Teacher(Person):
    def __init__(self, name: str, age: int, ls_of_studs=[]):
        super().__init__(name, age)
        self.ls_of_studs = ls_of_studs

    def teach(self):
        for student in self.ls_of_studs:
            student.get_knowledge()


class Assistant(Teacher, Student):
    def __init__(self, name: str, age: int, ls_of_studs, year=1):
        Teacher.__init__(self, name, age, ls_of_studs)
        Student.__init__(self, name, age, year)


print("Person:")
persona = Person("Oleg", 98)
print(f"Name: {persona.name}, age: {persona.age}")

print("\nStudent 47:")
student47 = Student("Sergey", 16)
print(f"Name: {student47.name}, age: {student47.age}, {
student47.name}'s knowledge: {student47.knowledge}")

student47.get_knowledge()
print(f"{student47.name}'s knowledge now {student47.knowledge}")

print("\nStudent 48:")
student48 = Student("Kate", 17)
print(f"Name: {student48.name}, age: {student48.age}, {student48.name}'s knowledge: {student48.knowledge}")

print("\nTeacher:")
teacher = Teacher("Tenigin A.A", 52, [student47, student48])
print(f"Name: {persona.name}, age: {persona.age}, list of students: {
      [stud.name for stud in teacher.ls_of_studs]}")

print("\nStuding:")
teacher.teach()
print(f"{student47.name} knowledge now {student47.knowledge}")
print(f"{student48.name} knowledge now {student48.knowledge}")

print("\nAssistant:")
assistant = Assistant("Olegsandrov", 25, [student47, student48])
print(f"Name: {assistant.name}, age: {assistant.age}, knowledge: {
      assistant.knowledge} list of students: {[stud.name for stud in assistant.ls_of_studs]}")

print("\nAssistant studing:")
assistant.teach()
print(f"{student47.name} knowledge now {student47.knowledge}")
print(f"{student48.name} knowledge now {student48.knowledge}")

print("\nAssistant get knowledge:")
assistant.get_knowledge()
print(f"{assistant.name} knowledge now {assistant.knowledge}")
assistant.get_knowledge()
print(f"{assistant.name} knowledge now {assistant.knowledge}")
```
