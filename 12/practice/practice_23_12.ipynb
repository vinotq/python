```python
def find_denom(n):
    for i in range(n):
        return n + [num for num in range(1, int(n**0.5)) if n % num == 0]


def proper_fractions(n):
    denom = [num for num in range(1, int(n**0.5)) if n % num != 0]
    return denom
```
