# **Динамическое пограммирование**
- способ решения задач, при котором одна большая задача разбивается на более мелкие (и желательно более понятные)


### **ДП сверху**
- *мемоизация*
- **пример** - нахождение чисел Фиббоначи
- *запоминание* результатов решения подзадач, которые можем встретить в дальнейшем

### **ДП снизу**
- *табуляция*
- переформулировка задачи в виде *рекурсии* последовательности простых подзадач

## **НОП**
### Для 
### {a} = [1, **2**, **3**, 4, **5**]
### и
### {b} = [**2**, 7, **3**, 2, **5**]
### ===>>>
## **НОП** = ***[2, 3, 5]***


## **О рюкзаке**
- Есть N предметов, **n_i** имеет вес **w_i > 0** и **p_i** (cost) **> 0**. Найти наибольший вариант для объема W


## **Расстояние Левенштейна**
- минимальное кол-во операций вставки, замены и удаления одного символа до индентичности обеих строк
