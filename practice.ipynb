# 1. Дано бинарное дерево. Выполнить прямой обход дерева.

## Ограничения:

$$
1<=n<=10^3
$$

## Пример 1:

### Ввод:

```py
'''
  2
 / \
1   3
'''
root = TreeNode(2, TreeNode(1), TreeNode(3))
```

### Вывод:

```py
[2, 1, 3]
```


````python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.data = val
        self.left = left
        self.right = right


def t3_20(root):
    if root is None:
        return []
    return [root.data] + t3_20(root.left) + t3_20(root.right)


assert t3_20(TreeNode(2, TreeNode(1), TreeNode(3))) == [2, 1, 3]
````

## Пример 2:

### Ввод:


### Вывод:

````python
def t3_3(n):
    if n < 2:
        return False
    if n == 2:
        return True

    for d in range(2, int(n**0.5) + 1):
        if n % d == 0:
            return False
    return True


assert t3_3(3)
assert t3_3(4) is False
````

# 3. Дан взвешенный ориентированный граф, содержащий `N` вершин и `M` ребер, представленный списком смежности. Составить матрицу инцидентности данного графа.

## Ограничения:

$1<=N<=10^3$

$0<=M<=N(N-1)$

## Пример 1:

### Ввод:

```py
N = 3
M = 2
adjL = {0: [(1, 5)], 1: [(2, 2)], 2: []}
```

### Вывод:

```py
[[0, -5],
 [-2, 5],
 [2, 0]
 ]
```


````python
def t3_29(n, m, adjL):
    edges = set()
    for i in adjL:
        for j, w in adjL[i]:
            edges.add((i, j, w))

    matrix = [[0] * m for _ in range(n)]

    for k, (i, j, w) in enumerate(edges):
        matrix[i][k] = -w
        matrix[j][k] = w

    return matrix


args = {"n": 3, "m": 2, "adjL": {0: [(1, 5)], 1: [(2, 2)], 2: []}}
assert t3_29(**args) == [[0, -5], [-2, 5], [2, 0]]
````

# 4. Вычислите `n`-й член последовательности, заданной формулами:

$$
a_{2n} = a_n ­+ a_{n-1}, \newline
a_{2n+1} = a_n – a_{n-1}, \newline
a_0 = a_1 = 1.
$$

## Ограничения:

$$
1 <= n < 10^6
$$

## Пример 1:

### Ввод:

```py
n = 2
```

### Вывод:

```py
2
```

## Пример 2:

### Ввод:

```py
n = 10
```

### Вывод:

```py
4
```


````python
def t3_4(n):
    dp = [0] * (n + 1)
    dp[0] = 1

    if n > 0:
        dp[1] = 1

    for i in range(2, n + 1):
        if i % 2 == 0:
            dp[i] = dp[i // 2] + dp[(i // 2) - 1]
        else:
            dp[i] = dp[i // 2] - dp[(i // 2) - 1]

    return dp[n]


assert t3_4(2) == 2
assert t3_4(10) == 4
assert t3_4(10000) == 137
````

# 7. Дано бинарное дерево. Выполнить уровневый обход дерева.

## Ограничения:

$0<=n<=10^4$

## Пример 1:

### Ввод:

```py
    1
   / \
  2   5
 / \   \
3   4   6
root = TreeNode(1, TreeNode(2, TreeNode(3), TreeNode(4)), TreeNode(5, None, TreeNode(6)))
```

### Вывод:

```py
[1, 2, 5, 3, 4, 6]
```


````python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.data = val
        self.left = left
        self.right = right


def t3_23(root):
    if not root:
        return []

    result = []
    queue = [root]

    while queue:
        cur = queue.pop(0)
        result.append(cur.data)

        if cur.left:
            queue.append(cur.left)
        if cur.right:
            queue.append(cur.right)

    return result


root = TreeNode(
    1, TreeNode(2, TreeNode(3), TreeNode(4)), TreeNode(5, None, TreeNode(6))
)
assert t3_23(root) == [1, 2, 5, 3, 4, 6]
````

# 12. Дано бинарное дерево. Найти ширину дерева. Под шириной дерева понимается наибольшее количество узлов на одном уровне.

## Ограничения:

$$
1<=n<=10^3
$$

## Пример 1:

### Ввод:

```py
'''
  2
 / \
1   3
'''
root = TreeNode(2, TreeNode(1), TreeNode(3))
```

### Вывод:

```py
2
```

## Пример 1:

### Ввод:

```py
'''
1
  \
   2
    \
     3
'''
root = TreeNode(1, right=TreeNode(2, right=TreeNode(3)))
```

### Вывод:

```py
1
```


````python
from collections import deque


class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.data = val
        self.left = left
        self.right = right


def t3_27(root):
    if not root:
        return 0

    max_width = 0
    queue = deque([root])

    while queue:
        level_size = len(queue)
        max_width = max(max_width, level_size)

        for _ in range(level_size):
            node = queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    return max_width


assert t3_27(TreeNode(2, TreeNode(1), TreeNode(3))) == 2
assert t3_27(TreeNode(1, right=TreeNode(2, right=TreeNode(3)))) == 1
````

## Ограничения:

$1<=n<=100$

$0<=m<=n(n-1)$

## Пример:

### Ввод:

```py
n = 3
incM = [[-1,  1,  0],
        [ 1,  0, -1],
        [ 0, -1,  1]]
```

### Вывод:

```py
{
  (0, 1),
  (2, 0),
  (1, 2)
}
```


````python
def t3_16(n, incM):
    edges = set()

    for j in range(len(incM[0])):
        st = -1
        end = -1

        for i in range(0, n):
            if incM[i][j] == -1:
                st = i
            elif incM[i][j] == 1:
                end = i

        if st != -1 and end != -1:
            edges.add((st, end))

    return edges


print(t3_16(3, [[-1, 1, 0], [1, 0, -1], [0, -1, 1]]))
assert t3_16(3, [[-1, 1, 0], [1, 0, -1], [0, -1, 1]]) == {(0, 1), (2, 0), (1, 2)}
````

# 21. Найти Z-функцию для строки.

## Ограничения:

$1<=N<=10^3$

## Пример 1:

### Ввод:

```py
s = "aaaaa"
```

### Вывод:

```py
[0, 4, 3, 2, 1]
```

## Пример 2:

### Ввод:

```py
s = "aaabaab"
```

### Вывод:

```py
[0, 2, 1, 0, 2, 1, 0]
```

## Пример 3:

### Ввод:

```py
s = "abacaba"
```

### Вывод:

```py
[0, 0, 1, 0, 3, 0, 1]
```


````python
def t4_33(s):
    n = len(s)
    z = [0] * n
    left = 0
    right = 0

    for i in range(1, n):
        if i < right:
            z[i] = min(right - i, z[i - left])

        while i + z[i] < n and s[z[i]] == s[i + z[i]]:
            z[i] += 1

        if i + z[i] > right:
            left = i
            right = i + z[i]

    return z


assert t4_33("aaaaa") == [0, 4, 3, 2, 1]
assert t4_33("aaabaab") == [0, 2, 1, 0, 2, 1, 0]
assert t4_33("abacaba") == [0, 0, 1, 0, 3, 0, 1]
````

````python
def t4_25(n, roads):
    adj_matrix = [[float("inf")] * n for _ in range(n)]

    for i, j, w in roads:
        adj_matrix[i][j] = w
        adj_matrix[j][i] = w

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if i == j:
                    adj_matrix[i][j] == 0

                adj_matrix[i][j] = min(
                    adj_matrix[i][j], adj_matrix[i][k] + adj_matrix[k][j]
                )

    return adj_matrix.index(min(adj_matrix, key=lambda x: sum(x)))


assert t4_25(3, [(0, 1, 3), (1, 2, 4)]) == 1
````

````python
def t4_19(s):
    n = len(s)
    st = set(list(s))
    count = sorted([[i, s.count(i)] for i in st], key=lambda x: -x[1])
    res = count[0][0]
    count[0][1] -= 1

    for i in range(1, n):
        if count[i % len(st)][1] and res[i - 1] != count[i % len(st)][0]:
            res += count[i % len(st)][0]
            count[i % len(st)][1] -= 1
        else:
            return "-1"
    return res


assert t4_19("aaabbbb") == "bababab"
assert t4_19("bbbb") == "-1"
````

````python
import heapq
import random


def heapify(arr, i, j=None):
    if not j:
        j = len(arr)
    n_idx = i
    if 2 * i + 1 < j and arr[i] < arr[2 * i + 1]:
        n_idx = 2 * i + 1

    if 2 * i + 2 < j and arr[n_idx] < arr[2 * i + 2]:
        n_idx = 2 * i + 2

    if n_idx != i:
        arr[i], arr[n_idx] = arr[n_idx], arr[i]
        heapify(arr, n_idx, j)


def make_heap(arr):
    non_leaf = len(arr) // 2 + 1
    for i in range(non_leaf, -1, -1):
        heapify(arr, i)


def sort_heap(arr):
    make_heap(arr)
    for i in range(len(arr) - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, 0, i)

    return arr


def t4_18(arr):
    # stupid approach (mine one)
    heapq.heapify(arr)
    return [heapq.heappop(arr) for _ in range(len(arr))]
    res = []
    while arr:
        res.append(heapq.heappop(arr))
    return res


assert sort_heap([1, 3, 2, 5, 4]) == [1, 2, 3, 4, 5]
for _ in range(100):
    arr = [random.randint(0, 999) for _ in range(random.randint(100, 1000))]
    assert sort_heap(arr.copy()) == sorted(arr)
````

За проезд по каждой дороге взимается некоторая пошлина. Найдите путь из города `А` в город `B` с минимальной величиной `S+P`, где `S` - сумма длин дорог пути, а `P` - сумма пошлин проезжаемых дорог. В качестве ответа верните `S+P` найденного пути.

## Пример 1:

### Ввод:


````python
def t4_26(n, roads, s, p, A, B):
    adj_list = {i: [] for i in range(n)}
    for (
        (i, j),
        si,
        pi,
    ) in zip(roads, s, p):
        adj_list[i].append((j, si + pi))
        adj_list[j].append((i, si + pi))

    used = [0] * n
    dist = [float("inf")] * n
    dist[A] = 0

    while True:
        u = None
        m = float("inf")
        for i in range(n):
            if not used[i] and dist[i] < m:
                u = i
                m = dist[i]

        if u is None:
            break

        used[u] = 1
        for v, w in adj_list[u]:
            if not used[v] and dist[v] > dist[u] + w:
                dist[v] = dist[u] + w

    if dist[B] == float("inf"):
        return None

    return dist[B]


assert t4_26(3, [(0, 1), (1, 2), (0, 2)], [5, 3, 10], [3, 2, 2], 0, 2) == 12
assert t4_26(4, [(0, 1), (1, 2), (0, 2)], [5, 3, 10], [3, 2, 2], 0, 3) == None
````

````python
def t4_8(arr1, arr2):
    res = [0 for _ in range(len(arr1) + len(arr2))]

    i = j = k = 0
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            res[k] = arr1[i]
            i += 1

        else:
            res[k] = arr2[j]
            j += 1

        k += 1

    while i < len(arr1):
        res[k] = arr1[i]
        i += 1
        k += 1

    while j < len(arr2):
        res[k] = arr2[j]
        j += 1
        k += 1

    return res


assert t4_8([0, 2, 4, 6], [1, 3, 5]) == [0, 1, 2, 3, 4, 5, 6]
````
