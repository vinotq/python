# 3. Дан взвешенный ориентированный граф, содержащий `N` вершин и `M` ребер, представленный списком смежности. Составить матрицу инцидентности данного графа.

## Ограничения:

$1<=N<=10^3$

$0<=M<=N(N-1)$

## Пример 1:

### Ввод:

```py
N = 3
M = 2
adjL = {0: [(1, 5)], 1: [(2, 2)], 2: []}
```

### Вывод:

```py
[[0, -5],
 [-2, 5],
 [2, 0]
 ]
```


````python
def t3_29(n, m, adjL):
    edges = set()
    for i in adjL:
        for j, w in adjL[i]:
            edges.add((i, j, w))

    matrix = [[0] * m for _ in range(n)]

    for k, (i, j, w) in enumerate(edges):
        matrix[i][k] = -w
        matrix[j][k] = w

    return matrix


args = {
    'n': 3,
    'm': 2,
    'adjL': {0: [(1, 5)], 1: [(2, 2)], 2: []}
}
assert t3_29(**args) == [[0, -5], [-2, 5], [2, 0]]
````

# 25.

Система двусторонних дорог такова, что для любой пары городов можно указать соединяющий их путь. Найдите такой город, сумма расстояний от которого до остальных городов минимальна.

На вход подается число городов `n`, список дорог `roads` размером `m`. Каждая запись в списке `roads` состоит из 3 чисел: `i` и `j` - номера городов начала и конца дороги, `s` - длина дороги. Каждый город пронумерован числом от `0` до `n-1`.

В качестве ответа верните номер города, сумма расстояний от которого до остальных городов минимальна. Если таких городов несколько, верните город с наименьшим номером.

## Ограничения:

$1<=n<=100$

$0<=m<=n(n-1)$

## Пример:

### Ввод:

```py
n = 3
roads = [(0, 1, 3), (1, 2, 4)]
```

### Вывод:

```py
1
```


````python
def t4_25(n, roads):
    adj_matrix = [[float('inf')] * n for _ in range(n)]

    for i, j, w in roads:
        adj_matrix[i][j] = w
        adj_matrix[j][i] = w

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if i == j:
                    adj_matrix[i][j] == 0

                adj_matrix[i][j] = min(
                    adj_matrix[i][j], adj_matrix[i][k] + adj_matrix[k][j])

    return adj_matrix.index(min(adj_matrix, key=lambda x: sum(x)))


assert t4_25(3, [(0, 1, 3), (1, 2, 4)]) == 1
````

Список должен использовать тот же класс Node, правый указатель должен ссылаться на следующий элемент в списке, левый всегда None.

Список должен иметь тот же порядок, что и прямой обход бинарного дерева.

Если дереве пустое верните `None`.

## Ограничения:

## Пример 1:

### Ввод:

```py
    1
   / \
  2   5
 / \   \
3   4   6
```

### Вывод:

### Пояснение:

После преобразования список выглядит следующим образом:

```py

     \
```

## Пример 2:

### Ввод:

```py
  1
 / \

   /

```

### Вывод:


### Пояснение:

После преобразования список выглядит следующим образом:

```py
1
 \
  3
   \
    4
     \
      2
       \
        5
```


````python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.data = val
        self.left = left
        self.right = right

    def __eq__(self, __value: object) -> bool:
        if self is None and __value is None:
            return True
        if self is None and __value is not None or \
           self is not None and __value is None:
            return False
        return self.data == __value.data and \
            self.left == __value.left and \
            self.right == __value.right


def preorder(root):
    if not root:
        return []
    return [root] + preorder(root.left) + preorder(root.right)


def t4_29(root):
    preorderer = preorder(root)
    for i in range(len(preorderer) - 1):
        preorderer[i].right = preorderer[i+1]
        preorderer[i].left = None

    preorderer[-1].left = None
    preorderer[-1].right = None

    return preorderer


root = TreeNode(1, TreeNode(2, TreeNode(3), TreeNode(4)),
                TreeNode(5, None, TreeNode(6)))
t4_29(root)
assert root == TreeNode(1, right=TreeNode(2, right=TreeNode(
    3, right=TreeNode(4, right=TreeNode(5, right=TreeNode(6))))))

root = TreeNode(1, TreeNode(3), TreeNode(4, TreeNode(2, right=TreeNode(5))))
t4_29(root)
assert root == TreeNode(1, right=TreeNode(
    3, right=TreeNode(4, right=TreeNode(2, right=TreeNode(5)))))
````

# 37.

За проезд по каждой дороге взимается некоторая пошлина. Найдите путь из города `А` в город `B` с минимальной величиной `S+P`, где `S` - сумма длин дорог пути, а `P` - сумма пошлин проезжаемых дорог. В качестве ответа верните `S+P` найденного пути.

На вход подается количество городов `n`, список дорог `roads` длиной `m`, список длин дорог `s` и список пошлин `p`, а так же города `А` и `B`. Все города пронумерованы от `0` до `n-1`. Все дороги двусторонние.

Если пути из `А` в `B` не существует верните `None`.

## Ограничения:

$1<=n<=100$

$1<=m<=n^2$

## Пример 1:

### Ввод:

```py
n = 3
roads = [(0, 1), (1, 2), (0, 2)]
s = [5, 3, 10]
p = [3, 2, 2]
A = 0
B = 2
```

### Вывод:

```py
12
```

## Пример 2:

### Ввод:

```py
n = 4
roads = [(0, 1), (1, 2), (0, 2)]
s = [5, 3, 10]
p = [3, 2, 2]
A = 0
B = 3
```

### Вывод:

```py
None
```


````python
def t4_26(n, roads, s, p, A, B):
    adj_list = {i: [] for i in range(n)}
    for (i, j), si, pi, in zip(roads, s, p):
        adj_list[i].append((j, si + pi))
        adj_list[j].append((i, si + pi))

    used = [0] * n
    dist = [float("inf")] * n
    dist[A] = 0

    while True:
        u = None
        m = float("inf")
        for i in range(n):
            if not used[i] and dist[i] < m:
                u = i
                m = dist[i]

        if u is None:
            break

        used[u] = 1
        for v, w in adj_list[u]:
            if not used[v] and dist[v] > dist[u] + w:
                dist[v] = dist[u] + w

    if dist[B] == float("inf"):
        return None

    return dist[B]


assert t4_26(3, [(0, 1), (1, 2), (0, 2)], [5, 3, 10], [3, 2, 2], 0, 2) == 12
assert t4_26(4, [(0, 1), (1, 2), (0, 2)], [5, 3, 10], [3, 2, 2], 0, 3) == None
````
